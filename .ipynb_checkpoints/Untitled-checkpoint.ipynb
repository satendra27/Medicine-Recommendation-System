```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Training.csv")
df.head()
```

```python
df.shape
```

```python
df["prognosis"].unique()
```

```python
len(df["prognosis"].unique())
```

```python
df.isnull().sum().sum()
```

```python
x = df.drop(columns=["prognosis"])
y = df["prognosis"]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
y_ar = le.fit_transform(y)
```

```python
y = pd.DataFrame(y_ar,columns=["prognosis"])
y.sample(3)
```

```python
y = np.ravel(y)
```

```python
y.ndim
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
models = {"LogisticRegression":LogisticRegression(),
          "SVC":SVC(),
          "RandomForest":RandomForestClassifier(),
          "Multinomial":MultinomialNB(),
          "DecisionTree":DecisionTreeClassifier(),
         }
```

```python
for model_name,model in models.items():
    model.fit(x_train,y_train)

    y_pred = model.predict(x_test)

    accuracy = accuracy_score(y_test,y_pred)

    cm = confusion_matrix(y_test,y_pred)
    print(f"{model_name}:{accuracy},\nConfusion Matrix:{cm}")
```

```python
svm = SVC()
svm.fit(x_train,y_train)
```

```python
y_pred = svm.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)*100
```

```python
import pickle
```

```python
pickle.dump(svm,open("svm_model.pkl",'wb'))
```

```python
# test 1
print("Predicted Label",svm.predict(x_test.iloc[0].values.reshape(1,-1)))
print("Actual Label",y_test[0])
```

```python
# test 2
print("Predicted Label",svm.predict(x_test.iloc[5].values.reshape(1,-1)))
print("Actual Label",y_test[5])
```
